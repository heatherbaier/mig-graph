```python
import geopandas as gpd
import pandas as pd
import json
```

```python
# Prep data
with open("../census_RAM/data/socio_vars/sum_num_intmig.json") as m:
    mig_data = json.load(m)
mig_data = pd.DataFrame.from_dict(mig_data, orient = 'index').reset_index()
mig_data.columns = ['muni_id', 'var']
"{:,}".format(mig_data['var'].sum()) + " total international migrants form Mexico in 2005-2010 census."
```

```python
df = pd.read_csv("../borderGraph/data/mexico_border_2010.csv")
"{:,}".format(df["num_mig"].sum()) + " total arrivals at each border station along US-Mexico border in 2010."
```

```python
bs = pd.read_csv("./data/tabula-U.S. Border Patrol Apprehensions From Mexico and Other Than Mexico (FY 2000 - FY 2020).csv")
bs = bs[["SECTOR", "FY 2005", "FY 2006", "FY 2007", "FY 2008", "FY 2009", "FY 2010"]]
for col in bs.columns:
    if col != "SECTOR":
        bs[col] = bs[col].str.replace(",", "").astype(float)
bs = bs.replace("Big Bend\r(formerly Marfa)", "Big Bend").replace("Rio Grande Valley\r(formerly McAllen)", "Rio Grande")
bs
```

```python
"{:,}".format(bs[bs["SECTOR"] == "Southwest Border"].drop(['SECTOR'], axis = 1).sum(axis = 1).squeeze()) + " total apprehensions at each Southwest border sector along US-Mexico border between 2005 and 2010."
```

```python
"{:,}".format(bs[bs["SECTOR"].isin(["Southwest Border", "Coastal Border"])].drop(['SECTOR'], axis = 1).sum(axis = 1).sum(axis = 0)) + " total apprehensions at each Southwest & Coastal border sector along US-Mexico border between 2005 and 2010."
```

```python
"{:,}".format(bs[bs["SECTOR"] == "Grand Total"].drop(['SECTOR'], axis = 1).sum(axis = 1).squeeze()) + " total apprehensions at all border sectors between 2005 and 2010."
```

```python
stations["sector"] = stations['portname'].map(port_map)
stations.head()
```

```python
bs = bs[bs['SECTOR'].isin(port_map.values())]
"{:,}".format(bs.sum(axis = 1).sum()) + " total apprehensions at each Southwest border sector along US-Mexico border between 2005 and 2010."
```

```python
bs_counts = pd.DataFrame(bs.sum(axis = 1))
bs_counts['sector'] = bs["SECTOR"]
bs_counts.columns = ["num_mig", "sector"]
bs_counts
```

```python
bs_counts.to_csv("./num_migrants_by_sector.csv", index = False)
```
